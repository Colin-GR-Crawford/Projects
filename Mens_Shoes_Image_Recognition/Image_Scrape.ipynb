```python
engine=create_engine('postgresql://localhost:5432/capstone_project')
```

```python
Data=pd.read_sql_table('mens_shoes_cleaned',con=engine)
```

```python
SAF=Data[Data.retailer=='Saks_Fifth_Avenue']
SAF['image_url_alt']=SAF['image_url']
SAF['image_url']=SAF['image_url'].apply(lambda x: x.replace('_A1_','_'))
```

```python
Output=pd.concat([Data[Data.retailer!='Saks_Fifth_Avenue'],SAF])
```

```python
Output=Output.ix[np.random.permutation(len(Output))].reset_index().drop('index',axis=1)
```
